```python
from sklearn import datasets
import matplotlib.pyplot as plt
```

```python
# Load the iris dataset
iris = datasets.load_iris()
```

```python
iris.data
```

```python
iris.target
```

```python
_, ax = plt.subplots()
scatter = ax.scatter(iris.data[:, 0], iris.data[:, 1], c=iris.target)
ax.set(xlabel=iris.feature_names[0], ylabel=iris.feature_names[1])
_ = ax.legend(
    scatter.legend_elements()[0], iris.target_names, title="Classes")
```

```python
from sklearn.decomposition import PCA

fig = plt.figure(1, figsize=(8, 6))
ax = fig.add_subplot(111, projection="3d", elev=-150, azim=110)

X_reduced = PCA(n_components=3).fit_transform(iris.data)
ax.scatter(
    X_reduced[:, 0],
    X_reduced[:, 1],
    X_reduced[:, 2],
    c=iris.target,
    s=40,
)

ax.set_title("First three PCA dimensions")
ax.set_xlabel("1st Eigenvector")
ax.xaxis.set_ticklabels([])
ax.set_ylabel("2nd Eigenvector")
ax.yaxis.set_ticklabels([])
ax.set_zlabel("3rd Eigenvector")
ax.zaxis.set_ticklabels([])

plt.show()
```

```python

fig = plt.figure(1, figsize=(8, 6))
ax = fig.add_subplot(111, projection="3d", elev=-150, azim=110)

ax.scatter(
    iris.data[:, 0],
    iris.data[:, 1],
    iris.data[:, 2],
    c=iris.target,
    s=40,
)

ax.set_xlabel(iris.feature_names[0])
ax.xaxis.set_ticklabels([])
ax.set_ylabel(iris.feature_names[1])
ax.yaxis.set_ticklabels([])
ax.set_zlabel(iris.feature_names[2])
ax.zaxis.set_ticklabels([])

plt.show()
```

```python
# Create figure with subplots - one for each target class
fig, axes = plt.subplots(1, 3, figsize=(15, 5))
fig.suptitle('Feature Distributions by Class')

# Loop through each target class
for i, target_name in enumerate(iris.target_names):
    # Get data for this class
    class_data = iris.data[iris.target == i]
    
    # Box plot for this class
    axes[i].boxplot(class_data)
    axes[i].set_title(f'Class: {target_name}')
    axes[i].set_xlabel('Features')
    axes[i].set_ylabel('Value')
    axes[i].set_xticklabels([name.split(' ')[0] for name in iris.feature_names])

plt.tight_layout(rect=[0, 0, 1, 0.95])  # Adjust layout to make room for the title
plt.show()
```

```python
# Create a figure with subplots - one for each feature
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle('Feature Distributions by Class', fontsize=16)
axes = axes.flatten()


# Loop through each feature
for i, feature_name in enumerate(iris.feature_names):
    # Plot histograms for each class
    for j, target_name in enumerate(iris.target_names):
        # Get data for this class and feature
        class_data = iris.data[iris.target == j][:, i]
        
        # Plot histogram
        axes[i].hist(class_data, alpha=0.5, bins=10, label=target_name)
    
    # Set titles and labels
    axes[i].set_title(feature_name)
    axes[i].set_xlabel('Value')
    axes[i].set_ylabel('Frequency')
    axes[i].legend()

plt.show()
```

```python
from sklearn.model_selection import train_test_split

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, random_state=1,stratify=iris.target,test_size=0.3)
```

```python
from sklearn import svm
clf = svm.SVC(kernel='linear',decision_function_shape='ovr')
clf.fit(X_train, y_train)
```

```python
predictions=clf.predict(X_test)
from sklearn.metrics import accuracy_score
print("Accuracy = ",accuracy_score(y_test, predictions))

from sklearn.metrics import confusion_matrix
from seaborn import heatmap
heatmap(confusion_matrix(y_test, predictions), annot=True, fmt='d', cmap='Blues')
```
